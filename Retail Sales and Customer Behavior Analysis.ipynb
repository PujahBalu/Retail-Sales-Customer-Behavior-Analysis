```python
# Import core data manipulation libraries
import pandas as pd
import numpy as np

# Import preprocessing tools from scikit-learn
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Set display options to easily view wide dataframes
pd.set_option('display.max_columns', None)

file_path = '/Users/pujahbalasubramaniam/Desktop/retail_data.csv' 

# Load the data directly
df = pd.read_csv(file_path, encoding='unicode_escape')
print(f"Data loaded successfully. Initial shape: {df.shape}")

# Display the first few rows
df.head()
```

```python
# Handle Missing Values and Anomalies

# Clean up column names by removing any hidden accidental spaces
df.columns = df.columns.str.strip()

# Drop rows where the customer ID is missing
df_clean = df.dropna(subset=['customer_id']).copy()

# Remove cancelled transactions (denoted by negative or zero quantities)
df_clean = df_clean[df_clean['quantity'] > 0]

# Ensure appropriate data types
df_clean['transaction_date'] = pd.to_datetime(df_clean['transaction_date'])

# Treat IDs as string categorical variables
df_clean['customer_id'] = df_clean['customer_id'].astype(int).astype(str) 
df_clean['transaction_id'] = df_clean['transaction_id'].astype(str)

print(f"Shape after cleaning: {df_clean.shape}")
```

```python
# Feature Engineering

# Calculate Total Spend per line item to ensure a clean metric for clustering
df_clean['Total_Spend'] = df_clean['quantity'] * df_clean['unit_price']

# Standardize product names (strip whitespace, upper case) for cleaner ARM results
df_clean['product_name'] = df_clean['product_name'].astype(str).str.strip().str.upper()

# Display our clean features
display(df_clean[['transaction_id', 'product_name', 'Total_Spend', 'transaction_hour']].head())
```

```python
# Preparation for Eclat (Association Rule Mining)
print("Preparing data for Eclat...")
basket = df_clean.groupby('transaction_id')['product_name'].apply(list).reset_index()
transactions_list = basket['product_name'].tolist()
print(f"Total distinct transactions for ARM: {len(transactions_list)}")


# --- RANDOM FOREST (Classification) ---
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("Preparing data and encoding categories...")

# --- THE MISSING SETUP: Define df_rf and create the Target Variable ---
df_rf = df_clean.copy()
threshold = df_rf['Total_Spend'].quantile(0.75)
df_rf['Is_High_Value'] = (df_rf['Total_Spend'] >= threshold).astype(int)
# ----------------------------------------------------------------------

# FOOLPROOF ENCODING: Guarantee no strings slip through
categorical_cols = ['store_location', 'day_of_week', 'month_of_year']
numeric_cols = ['quantity', 'unit_price', 'transaction_hour']

# Build X and Y safely
X_raw = df_rf[numeric_cols + categorical_cols]
X_class_encoded = pd.get_dummies(X_raw, columns=categorical_cols, drop_first=True, dtype=int)
y_target = df_rf['Is_High_Value']

# Save the exact feature names for our chart
rf_features = X_class_encoded.columns.tolist()

# Split the data (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X_class_encoded, y_target, test_size=0.2, random_state=42)

# Define the Model Evaluation Function
def evaluate_rf(max_depth_val, variation_name):
    print(f"\n--- Random Forest {variation_name} (Max Depth: {max_depth_val}) ---")
    rf_model = RandomForestClassifier(n_estimators=100, max_depth=max_depth_val, random_state=42, n_jobs=-1)
    rf_model.fit(X_train, y_train)
    
    y_pred = rf_model.predict(X_test)
    print(classification_report(y_test, y_pred))
    return rf_model

# Variation 1: Shallow Tree (Prevents overfitting, focuses on macro trends)
rf_model_1 = evaluate_rf(max_depth_val=5, variation_name="Variation 1")

# Variation 2: Deep Tree (Captures more complex purchasing behaviors)
rf_model_2 = evaluate_rf(max_depth_val=15, variation_name="Variation 2")

# Visualize Feature Importance from the deeper model
print("\nGenerating Feature Importance Chart...")
# Get the top 10 most important features
importances = pd.Series(rf_model_2.feature_importances_, index=rf_features).sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Updated seaborn plot without the deprecated legend argument
ax = sns.barplot(
    x=importances.values, 
    y=importances.index, 
    hue=importances.index, 
    palette="viridis"
)

# Safely remove the legend
ax.get_legend().remove()

plt.title("Random Forest: Top 10 Drivers of 'High-Value' Transactions")
plt.xlabel("Importance Score")
plt.ylabel("Purchasing Features")
plt.tight_layout()
plt.show()
```

```python
import time

print("Starting Eclat (Vertical Data Format) Analysis...")
start_time = time.time()

# Step A: Convert to Vertical Data Format (Item -> Set of Transaction IDs)
vertical_db = {}
for tid, transaction in enumerate(transactions_list):
    # Remove duplicates within the same transaction to prevent skewed support
    unique_items = set(transaction) 
    for item in unique_items:
        if item not in vertical_db:
            vertical_db[item] = set()
        vertical_db[item].add(tid)

total_transactions = len(transactions_list)

# Step B: Define a function to find frequent pairs via intersection
def run_eclat_pairs(min_support_pct):
    min_support_count = total_transactions * min_support_pct
    frequent_items = {item: tids for item, tids in vertical_db.items() if len(tids) >= min_support_count}
    frequent_pairs = {}
    items = list(frequent_items.keys())
    
    # Intersect the Transaction ID (TID) sets to find co-occurrences
    for i in range(len(items)):
        for j in range(i + 1, len(items)):
            item1, item2 = items[i], items[j]
            # The core of Eclat: intersection of TID sets
            intersection = frequent_items[item1].intersection(frequent_items[item2]) 
            if len(intersection) >= min_support_count:
                frequent_pairs[(item1, item2)] = len(intersection)
                
    # Sort by highest frequency
    return sorted(frequent_pairs.items(), key=lambda x: x[1], reverse=True)

# Variation 1: Strict Minimum Support (e.g., 2%)
print("\n--- Eclat Variation 1: 2% Minimum Support ---")
top_pairs_strict = run_eclat_pairs(0.02)
for pair, count in top_pairs_strict[:5]:
    print(f"Bundle: {pair} | Transactions: {count}")
# Variation 2: Loose Minimum Support (e.g., 1%) to capture more niche bundles
print("\n--- Eclat Variation 2: 1% Minimum Support ---")
top_pairs_loose = run_eclat_pairs(0.01)
for pair, count in top_pairs_loose[:5]:
    print(f"Bundle: {pair} | Transactions: {count}")

print(f"\nEclat execution completed in {time.time() - start_time:.2f} seconds.")
```

```python
# --- 2. RANDOM FOREST (Classification) - sendhalps ---
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("Executing foolproof preprocessing...")

# 1. Pull a completely fresh copy from our clean data to avoid memory glitches
rf_data = df_clean.copy()

# Define our target variable
threshold = rf_data['Total_Spend'].quantile(0.75)
y_target = (rf_data['Total_Spend'] >= threshold).astype(int)

# 2. Define our columns
categorical_cols = ['store_location', 'day_of_week', 'month_of_year']
numeric_cols = ['quantity', 'unit_price', 'transaction_hour']

# Isolate the features
X_raw = rf_data[numeric_cols + categorical_cols]

# 3. FORCE One-Hot Encoding and FORCE the output to integers (0 and 1)
# The dtype=int argument prevents newer versions of Pandas from creating weird boolean types
X_encoded = pd.get_dummies(X_raw, columns=categorical_cols, drop_first=True, dtype=int)

# --- SAFETY CHECK ---
print("\nVerifying data types before training (Should be empty if clean):")
print("Remaining text columns:", X_encoded.select_dtypes(include=['object', 'string']).columns.tolist())
# --------------------

# Save the exact feature names for our chart
rf_features = X_encoded.columns.tolist()

# 4. Split the data
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_target, test_size=0.2, random_state=42)

# 5. Define the Model Evaluation Function
def evaluate_rf(max_depth_val, variation_name):
    print(f"\n--- Random Forest {variation_name} (Max Depth: {max_depth_val}) ---")
    rf_model = RandomForestClassifier(n_estimators=100, max_depth=max_depth_val, random_state=42, n_jobs=-1)
    rf_model.fit(X_train, y_train) # <--- If it passes this, we are golden
    
    y_pred = rf_model.predict(X_test)
    print(classification_report(y_test, y_pred))
    return rf_model

# Run Variation 1: Shallow Tree
rf_model_1 = evaluate_rf(max_depth_val=5, variation_name="Variation 1")

# Run Variation 2: Deep Tree
rf_model_2 = evaluate_rf(max_depth_val=15, variation_name="Variation 2")

# 6. Visualize Feature Importance
print("\nGenerating Feature Importance Chart...")
# Get the top 10 most important features
importances = pd.Series(rf_model_2.feature_importances_, index=rf_features).sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))

# Removed the deprecated 'legend=False' argument
ax = sns.barplot(
    x=importances.values, 
    y=importances.index, 
    hue=importances.index, 
    palette="viridis"
)

# Disable the legend using matplotlib directly instead
ax.get_legend().remove()

plt.title("Random Forest: Top 10 Drivers of 'High-Value' Transactions")
plt.xlabel("Importance Score")
plt.ylabel("Purchasing Features")
plt.tight_layout()
plt.show()
```

```python
# --- 3. DBSCAN (Clustering) ---
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("Preparing customer metrics for clustering...")

# 1. Prepare the Data: Aggregate to the Customer level
customer_metrics_full = df_clean.groupby('customer_id').agg(
    Total_Transactions=('transaction_id', 'nunique'),
    Total_Volume=('quantity', 'sum'),
    Total_Revenue=('Total_Spend', 'sum')
).reset_index()

# --- THE FIX: RANDOM SAMPLING ---
# We take a random sample of 50,000 customers to prevent Kaggle's RAM from crashing
sample_size = min(50000, len(customer_metrics_full))
customer_metrics = customer_metrics_full.sample(n=sample_size, random_state=42).copy()
print(f"Sampled customer metrics down to {sample_size} records to prevent Out-Of-Memory errors.")

# 2. Scale the Data: DBSCAN strictly requires normalized continuous features
scaler = StandardScaler()
X_cluster = scaler.fit_transform(customer_metrics[['Total_Transactions', 'Total_Volume', 'Total_Revenue']])
print(f"Clustering features shape: {X_cluster.shape}")

# 3. Define the DBSCAN execution function
def run_dbscan(eps_val, min_samples_val, variation_name):
    print(f"\n--- DBSCAN {variation_name} (eps: {eps_val}, min_samples: {min_samples_val}) ---")
    
    # Initialize and fit the model
    dbscan = DBSCAN(eps=eps_val, min_samples=min_samples_val, n_jobs=-1)
    clusters = dbscan.fit_predict(X_cluster)
    
    # Calculate results
    n_clusters = len(set(clusters)) - (1 if -1 in clusters else 0)
    n_noise = list(clusters).count(-1)
    
    print(f"Estimated number of distinct customer segments: {n_clusters}")
    print(f"Number of anomaly/outlier customers (Noise): {n_noise}")
    return clusters

# Run Variation 1: Tight clustering
clusters_1 = run_dbscan(eps_val=0.3, min_samples_val=5, variation_name="Variation 1")

# Run Variation 2: Broader clustering 
clusters_2 = run_dbscan(eps_val=0.6, min_samples_val=5, variation_name="Variation 2")

# 4. Visualization
print("\nGenerating Customer Segments Chart...")
# Append the broader clusters to our customer metrics for visualization
customer_metrics['Cluster'] = clusters_2

plt.figure(figsize=(10, 6))
# A cluster value of -1 represents noise/outliers in DBSCAN
sns.scatterplot(
    data=customer_metrics, 
    x='Total_Volume', 
    y='Total_Revenue', 
    hue='Cluster', 
    palette='Set1', 
    legend='full',
    alpha=0.6
)
plt.title("DBSCAN Customer Segments (Revenue vs. Volume) - Sampled Data")
plt.xlabel("Total Units Purchased (Volume)")
plt.ylabel("Total Lifetime Spend (Revenue)")
plt.tight_layout()
plt.show()
```

```python
# --- Task 7: Final Executive Visualization ---
import matplotlib.pyplot as plt
import seaborn as sns

print("Generating Final Revenue Distribution Analysis...")

# Group our sampled customers by their exact volume tier to find the Average Spend
tier_revenue = customer_metrics.groupby('Total_Volume')['Total_Revenue'].mean().reset_index()
tier_revenue.rename(columns={'Total_Revenue': 'Average_Spend'}, inplace=True)

# Visualize the value of moving a customer up a single tier vs. substituting premium items
plt.figure(figsize=(10, 5))
sns.barplot(
    data=tier_revenue, 
    x='Total_Volume', 
    y='Average_Spend', 
    palette='mako'
)

plt.title("Executive Summary: Average Customer Spend per Volume Tier")
plt.xlabel("Exact Units Purchased (The Volume Pillars)")
plt.ylabel("Average Lifetime Spend")
plt.tight_layout()
plt.show()

# Calculate the Average Unit Retail (AUR) to highlight the premiumization gap
aur = df_clean['Total_Spend'].sum() / df_clean['quantity'].sum()
print(f"Overall Average Unit Retail (AUR): ${aur:.2f}")
```
